```python
# Load stops.txt -> export stops.geojson

stops_df = pd.read_csv(f'{FILEPATH}stops.txt')
df = stops_df

stops_geometry = geopandas.points_from_xy(stops_df.stop_lon, stops_df.stop_lat)
stops_geo_df = geopandas.GeoDataFrame(df[['stop_id','stop_name']], geometry=stops_geometry)

stops_geo_df.to_file("stops.geojson", driver='GeoJSON')
```
